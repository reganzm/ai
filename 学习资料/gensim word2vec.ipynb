```python
import sys
import os
import pandas as pd
f = open('wiki.zh.word.text','r',encoding='utf-8')
data = f.readline()
df = []
error = 0
right = 0
while True:
    try:
        line = f.readline()
        if line:
            df.append(line)
            right += 1
        else:
            break
    except Exception as e:
        error += 1
        print(e)
print('error:%s,right:%s'%(error,right))
```

```python
with open('wiki.zh.word.text.utf8','a+',encoding='utf-8') as f:
    for data in df:
        f.write(data+'\n')
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

# 
sentences = LineSentence('wiki.zh.word.text.utf8')
# size
# window
# min_countmin_count
model = Word2Vec(sentences, size=128, window=5, min_count=5, workers=4)

# 
model.save('word_embedding_128')

# 
# 
# model = Word2Vec.load("word_embedding_128")

# 
# 
items = model.most_similar(u'')
for item in items:
    # 
    print (item[0], item[1])
# 
model.similarity(u'',  u'')
```
