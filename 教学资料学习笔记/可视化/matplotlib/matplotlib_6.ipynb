```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
X = np.arange(-10,10,0.1)
```

```python
#对X使用Sigmod函数求求值
Y = list(map(lambda x:1/(1+math.pow(math.e,-x)),X))
```

```python
#绘图
plt.plot(X,Y,'r-.')
plt.show()
```

```python
plt.plot(X,Y,color='b',linestyle='--')
plt.show()
```

```python
plt.plot(X,Y,color='#FFCE44',linestyle='-.')
plt.show()
```

```python
#绘制笛卡尔坐标系
plt.plot(X,Y,color='#93349D',linestyle='-.')
#得到Axes对象，用于操作坐标轴
ax = plt.gca()
#设置上、右边框颜色为none，达到删除的目的
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#设置x轴上标度的位置，位于坐标轴下方
ax.xaxis.set_ticks_position('bottom')
#设置x轴中心位置为0点
ax.spines['bottom'].set_position(('data',0))
#设置y轴上标度的位置，位于坐标轴左边
ax.yaxis.set_ticks_position('left')
#设置y轴中心位置为0点
ax.spines['left'].set_position(('data',0))
plt.show()
```
