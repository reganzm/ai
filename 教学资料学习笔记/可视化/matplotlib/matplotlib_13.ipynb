```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig = plt.figure()
#*left*, *bottom*, *width*,*height*
ax = fig.add_axes([0.,0.,1.,1.])
inner_ax = fig.add_axes([0.6,0.6,0.3,0.3])
plt.show()
```

```python
#位两个坐标产生数据
x1 = np.arange(10)
y1 = np.random.randint(0,10,10)

x2 = np.arange(5)
y2 = np.random.randint(0,10,5)

fig = plt.figure()
#*left*, *bottom*, *width*,*height*
ax = fig.add_axes([0.,0.,1.,1.])
inner_ax = fig.add_axes([0.7,0.7,0.25,0.25])

ax.plot(x1,y1)
ax.grid()
inner_ax.plot(x2,y2)

plt.show()
```

```python
#子图网格
import matplotlib.pyplot as plt
gs = plt.GridSpec(2,2)
fig = plt.figure(figsize=(10,10))
fig.add_subplot(gs[0,0],title='a')
fig.add_subplot(gs[1,0],title='b')
fig.add_subplot(gs[:2,1],title='c')
plt.show()
```

```python
import matplotlib.pyplot as plt
gs = plt.GridSpec(2,2)
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(gs[0,0],title='a')
ax1.pie([4,5,6,7,8],labels=['a','b','c','d','e'])
ax2 = fig.add_subplot(gs[1,0],title='b')
ax2.barh([2,3,4,6],[6,7,9,12])
ax3 = fig.add_subplot(gs[:2,1],title='c')
ax3.plot(np.arange(10),np.random.randint(10,20,10))
plt.show()
```
