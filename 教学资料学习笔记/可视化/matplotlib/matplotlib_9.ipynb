```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
datas = [33,16,56,90,55]
colors = ['red','green','blue','pink','yellow']
labels = ['xiaomi','huawei','apple','oppo','vivo']
plt.pie(datas,colors=colors,labels=labels)
plt.show()
```

```python
#正圆
datas = [33,16,56,90,55]
colors = ['red','green','blue','pink','yellow']
labels = ['xiaomi','huawei','apple','oppo','vivo']
plt.pie(datas,colors=colors,labels=labels)
plt.axis('equal')
plt.show()
```

```python
#脱离
datas = [33,16,56,90,55]
colors = ['red','green','blue','pink','yellow']
labels = ['xiaomi','huawei','apple','oppo','vivo']
explode = [0.2,0,0,0,0]
plt.pie(datas,colors=colors,labels=labels,explode=explode)
plt.axis('equal')
plt.show()
```

```python
#标题和旋转
datas = [33,16,56,90,55]
colors = ['red','green','blue','pink','yellow']
labels = ['xiaomi','huawei','apple','oppo','vivo']
explode = [0.2,0,0,0,0]
plt.pie(datas,colors=colors,labels=labels,explode=explode,startangle=60)
plt.axis('equal')
plt.title('pie chart')
plt.show()
```

```python
#autopct自动计算百分比并显示在扇形上
datas = [33,16,56,90,55]
colors = ['red','green','blue','pink','yellow']
labels = ['xiaomi','huawei','apple','oppo','vivo']
explode = [0.2,0,0,0,0]
plt.pie(datas,colors=colors,labels=labels,explode=explode,startangle=180,autopct='%1.1f%%')
plt.axis('equal')
plt.title('pie chart')
plt.show()
```

```python
#添加阴影效果
datas = [33,16,56,90,55]
colors = ['red','green','blue','pink','yellow']
labels = ['xiaomi','huawei','apple','oppo','vivo']
explode = [0.2,0,0,0,0]
plt.pie(datas,colors=colors,labels=labels,explode=explode,startangle=180,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.title('pie chart')
plt.show()
```
