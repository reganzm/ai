```python
## numpy,
import math
import numpy as np
data = np.arange(1,5,0.1)
data
```

```python
y1 = list(map(math.cos,math.pi * data))
y2 = list(map(math.cos,math.pi * data + math.pi / 2))
y3 = list(map(math.cos,math.pi * data - math.pi / 2))
```

```python
#data,y1,y2,y3
plt.plot(data,y1,'bo',data,y2,'r-.',data,y3,'gs')
plt.show()
```

```python
#data,y1,y2,y3
plt.plot(data,y1,'bo--',data,y2,'r--.',data,y3,'gs-')
plt.show()
```
