```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成10个0到10之间的整数
data = np.random.randint(0,10,10)
data
```

```python
x = np.arange(10)
#绘制条形图
plt.bar(x,data)
plt.show()
```

```python
#水平条形图
plt.figure(figsize=(20,10))
plt.barh(x,data,label='bar')
plt.legend(loc=1)
plt.title('horizontal bar chart')

plt.show()
```

```python
#多序列条形图
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
index = np.arange(5)
data1 = [3,4,6,8,9]
data2 = [11,23,6,5,1]
data3 = [12,6,21,8,26]
#把一个空间分为3部分，定义alpha值为0.3，三个占0.9，剩0.1的空格
alpha = 0.3
plt.title('multi bar chart')
plt.bar(index,data1,alpha,color='r',label='a')
plt.bar(index+alpha,data2,alpha,color='b',label='b')
plt.bar(index+2*alpha,data3,alpha,color='k',label='c')
plt.legend(loc=2)
plt.show()
```

```python
#多序列条形图转换为堆积条形图
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
index = np.arange(5)
data1 = np.array([3,4,6,8,9])
data2 = np.array([11,23,6,5,1])
data3 = np.array([12,6,21,8,26])
#把一个空间分为3部分，定义alpha值为0.3，三个占0.9，剩0.1的空格
plt.bar(index,data1,color='r',label='a')
plt.bar(index,data2,bottom=data1,color='b',label='b')
plt.bar(index,data3,bottom=(data2+data1),color='k',label='c')
plt.legend(loc=2)
plt.show()
```

```python
#多序列条形图转换为堆积条形图
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
index = np.arange(5)
data1 = np.array([3,4,6,8,9])
data2 = np.array([11,23,6,5,1])
data3 = np.array([12,6,21,8,26])
#把一个空间分为3部分，定义alpha值为0.3，三个占0.9，剩0.1的空格
plt.barh(index,data1,color='r',label='a')
plt.barh(index,data2,left=data1,color='b',label='b')
plt.barh(index,data3,left=(data2+data1),color='k',label='c')
plt.legend()
plt.show()
```

```python
#多序列条形图转换为堆积条形图,使用填充线
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
index = np.arange(5)
data1 = np.array([3,4,6,8,9])
data2 = np.array([11,23,6,5,1])
data3 = np.array([12,6,21,8,26])
#把一个空间分为3部分，定义alpha值为0.3，三个占0.9，剩0.1的空格
plt.barh(index,data1,color='w',label='a',hatch='xxxxx')
plt.barh(index,data2,left=data1,color='w',label='b',hatch='////')
plt.barh(index,data3,left=(data2+data1),color='w',label='c',hatch='\\\\\\\\')
plt.legend(loc=4)
plt.show()
```

```python
#pandas DataFrame绘制堆积条形图
df = pd.DataFrame([[4,8,8,2],[4,9,1,3],[4,8,8,1]],columns=['a','b','c','d'])
print(df.T)
df.T.plot(kind='bar',stacked=True)
```
