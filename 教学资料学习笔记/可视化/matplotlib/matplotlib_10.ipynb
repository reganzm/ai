```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#1.生成表格图
x = np.linspace(-2,2,500)
y = np.linspace(-2,2,500)
X,Y = np.meshgrid(x,y)
```

```python
#定义z函数,L2正则化
def z(x,y):
    return  x**2 + y**2
#定义函数f,L1正则化
def f(x,y):
    return np.abs(x)+np.abs(y)
```

```python
#参数分别是X轴网格线，Y轴网格线，XY对应值求出的z值，等高线个数为5个，颜色关键字参数
#contour用于设置等高线颜色个数等属性
C = plt.contour(X,Y,z(X,Y),5)
#contourf设置图表区域，参数和contour一样
plt.contourf(X,Y,z(X,Y),5,cmap=plt.cm.hot)
#使用C设置线上的标签值，设置字体大小等
plt.clabel(C,inline=1,fontsize=10)
plt.show()
```

```python
#参数分别是X轴网格线，Y轴网格线，XY对应值求出的z值，等高线个数为5个，颜色关键字参数
#contour用于设置等高线颜色个数等属性
C = plt.contour(X,Y,f(X,Y),5,color='black')
#contourf设置图表区域，参数和contour一样
plt.contourf(X,Y,f(X,Y),5,cmap=plt.cm.hot)
#使用C设置线上的标签值，设置字体大小等
plt.clabel(C,inline=1,fontsize=10)
plt.show()
```
