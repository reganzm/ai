```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#绘制3D曲面
#L2正则化
def L2(w1,w2):
    return w1**2 + w2**2
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)
X,Y = np.meshgrid(x,y)
ax.plot_surface(X,Y,L2(X,Y))
plt.show()
```

```python
#再尝试一下L1正则项
def L1(w1,w2):
    return np.abs(w1)+np.abs(w2)
fig = plt.figure()
ax = Axes3D(fig)
x = np.linspace(1,5,50)
y = np.linspace(1,5,50)
X,Y = np.meshgrid(x,y)
ax.plot_surface(X,Y,L1(X,Y))
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)
X,Y = np.meshgrid(x,y)
ax.plot_surface(X,Y,L2(X,Y),cmap=plt.cm.hot)
ax.view_init(azim=20,elev=60)
plt.show()
```
