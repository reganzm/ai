```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
#绘制3D散点图
x1 = np.random.randint(10,20,100)
y1 = np.random.randint(10,20,100)
z1 = np.random.randint(10,20,100)
```

```python
#得到figure图表对象
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x1,y1,z1)
plt.show()
```

```python
x1 = np.random.randint(10,20,10)
y1 = np.random.randint(10,20,10)
z1 = np.random.randint(10,20,10)

x2 = np.random.randint(20,30,10)
y2 = np.random.randint(20,30,10)
z2 = np.random.randint(20,30,10)

x3 = np.random.randint(40,50,10)
y3 = np.random.randint(40,50,10)
z3 = np.random.randint(40,50,10)
#得到figure图表对象
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x1,y1,z1)
ax.scatter(x2,y2,z2,c='r',marker="*")
ax.scatter(x3,y3,z3,c='g',marker='s')
ax.set_xlabel('X label')
ax.set_ylabel('Y_label')
ax.set_zlabel('Z_label')
plt.show()
```
