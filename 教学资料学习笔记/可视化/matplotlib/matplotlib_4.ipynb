```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.linspace(1,5,100)
```

```python
#求2次方
data2 = pow(data,2)
```

```python

#绘制图表
plt.plot(data,data2,'y--')
#设置标题、颜色、字体、大小
plt.title('square',fontsize=25,fontname='Times New Roman',color='blue')
#设置x轴标签
plt.xlabel('X',fontsize=18,fontname='Times New Roman',color='red')
#设置y轴标签
plt.ylabel('Y',fontsize=12,fontname='Times New Roman',color='green')
plt.show()
```

```python
#绘制图表
plt.plot(data,data2,'y--')
#设置标题、颜色、字体、大小
plt.title('square',fontsize=25,fontname='Times New Roman',color='blue')
#设置x轴标签
plt.xlabel('X',fontsize=18,fontname='Times New Roman',color='red')
#设置y轴标签
plt.ylabel('Y',fontsize=12,fontname='Times New Roman',color='green')
#在（3,10）位置插入文本“square”
plt.text(3,10,'square',fontsize=20,color='green')
plt.show()
```

```python
#添加网格
#绘制图表
plt.plot(data,data2,'y--')
#设置标题、颜色、字体、大小
plt.title('square',fontsize=25,fontname='Times New Roman',color='blue')
#设置x轴标签
plt.xlabel('X',fontsize=18,fontname='Times New Roman',color='red')
#设置y轴标签
plt.ylabel('Y',fontsize=12,fontname='Times New Roman',color='green')
#在（3,10）位置插入文本“square”
plt.text(3,10,'square',fontsize=20,color='green')
#数学表达式
plt.text(3,23,r'$y=x^2$',fontsize=20,color='green')
#网格
plt.grid(True)
#图例
plt.legend([r'$y=x^2$'],loc=2)
plt.show()
```

```python
data3 = np.power(data,2.2)
#多序列图表
#绘制图表
plt.plot(data,data2,'y--')
#设置标题、颜色、字体、大小
plt.title('square',fontsize=25,fontname='Times New Roman',color='blue')
#设置x轴标签
plt.xlabel('X',fontsize=18,fontname='Times New Roman',color='red')
#设置y轴标签
plt.ylabel('Y',fontsize=12,fontname='Times New Roman',color='green')
#在（3,10）位置插入文本“square”
plt.text(3,10,'square',fontsize=20,color='green')
#数学表达式
plt.text(3,23,r'$y=x^2$',fontsize=20,color='green')
#网格
plt.grid(True)
#添加data3
plt.plot(data,data3,'r-.')
#图例
plt.legend([r'$y=x^2$',r'$y=x^(2.2)$'])
plt.show()
```

```python
#保存图片
data3 = np.power(data,2.2)
#多序列图表
#绘制图表
plt.plot(data,data2,'y--')
#设置标题、颜色、字体、大小
plt.title('square',fontsize=25,fontname='Times New Roman',color='blue')
#设置x轴标签
plt.xlabel('X',fontsize=18,fontname='Times New Roman',color='red')
#设置y轴标签
plt.ylabel('Y',fontsize=12,fontname='Times New Roman',color='green')
#在（3,10）位置插入文本“square”
plt.text(3,10,'square',fontsize=20,color='green')
#数学表达式
plt.text(3,23,r'$y=x^2$',fontsize=20,color='green')
#网格
plt.grid(True)
#添加data3
plt.plot(data,data3,'r-.')
#图例
plt.legend([r'$y=x^2$',r'$y=x^(2.2)$'])
plt.savefig('matplotlib.png')
plt.show()
```
