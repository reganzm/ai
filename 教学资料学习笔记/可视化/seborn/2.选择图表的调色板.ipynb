```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#使用sb上的set方法进行全局设置，通过figure.figsize设置图表的大小
sb.set(rc={"figure.figsize": (10, 6)})
#设置随机种子，便于结果复现
np.random.seed(sum(map(ord, "选择图表调色板")))
print(sum(map(ord, "选择图表调色板")))
```

```python
#color_palette方法返回默认的调色板信息
current_palette = sb.color_palette()
#实际上就是RGB的三色元组
print(current_palette)
sb.palplot(current_palette)
```

```python
#默认的调色板只有6中颜色，多余6中颜色的类别数据怎么设置颜色呢？
#做常见的方法就是使用hls颜色空间，第二个参数是产生的颜色的个数
sb.palplot(sb.color_palette("hls", 10))
```

```python
#也可是使用cubehelix方法
sb.cubehelix_palette(15)
```

```python
sb.palplot(sb.cubehelix_palette(15))
```

```python
#可以通过start设置颜色的开始位置,start取值范围是【0~3】，用rot设置旋转角度取值范围是[-1,1]
sb.palplot(sb.cubehelix_palette(15, start=3, rot=-1))
```

```python
#还可以控制颜色的亮度，使用dark设置最低的亮度，light设置最高的亮度,reverse设置反转
sb.palplot(sb.cubehelix_palette(15,start=0.5,rot=-1,dark=0.1,light=0.8,reverse=True))
```

```python
#调用调色板方法返回的是一组颜色的序列，可以用这个序列构建colormap对象，供seaborn或matplotlib使用。
#返回colormap对象只需要设置关键字参数as_cmap=True即可
```

```python
#多元正态分布，第一个参数为两个方向的维度的均值，第二个为两组数据的协方差矩阵需要时对称的半正定矩阵
x, y = np.random.multivariate_normal([1, 1], [[1, -.7], [-.5, 1]], size=300).T
cmap = sb.cubehelix_palette(light=1, as_cmap=True)
print(type(cmap))
sb.kdeplot(x, y, cmap=cmap, shade=True)
```

### 设置seaborn默认的颜色面板

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
#set_palette方法参数可能取值如下
sb.set_palette("hot")
sinplot()
```
