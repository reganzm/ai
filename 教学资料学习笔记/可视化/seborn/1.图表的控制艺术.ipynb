```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

### 图表风格

```python
print(help(sb.set_style))
```

```python
sb.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
print(data.shape)
sb.boxplot(data=data)
```

#### 有六种风格darkgrid, whitegrid, dark, white,  ticks，默认的风格是：darkgrid

```python
sb.set_style("ticks")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sb.boxplot(data=data)
```

```python
sb.set_style("white")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sb.boxplot(data=data)
```

### 去除图表边框

```python
#使用seaborn的despine方法去除边框
sb.set_style("ticks")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sb.boxplot(data=data)
sb.despine()
```

```python
sb.violinplot(data=data)
sb.despine()
```

```python
#通过参数，去除左边的轴
sb.set_style('whitegrid')
sb.violinplot(data=data)
sb.despine(left=True)
```

```python
#使用with语句设置临时风格
with sb.axes_style("dark"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(3)
```

```python
#通过set_style方法重写seaborn默认的风格元素 [However, the higher-level set() function takes a dictionary of any matplotlib parameters).]
#调用axis_style方法查看默认的style风格配置，他是一个字典结构
sb.axes_style()
```

### 图表元素的缩放

```python
#使用sb.set()方法重置默认参数
sb.set()
```

```python
#sb里面有缩放的四个级别的设置，分别是paper, notebook, talk, and poster，notebook是默认的
sb.set_context("paper")
sinplot()
```

```python
sb.set_context("notebook")
sinplot()
```

```python
sb.set_context("talk")
sinplot()
```

```python
sb.set_context("poster")
sinplot()
```

```python
#当然也可以对某种缩放风格的默认设置进行更该，调用set_context方法，传入一个字典结构即可
sb.set_context("poster", font_scale=1.5, rc={"lines.linewidth": 5.5})
sinplot()
```
