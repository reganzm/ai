```python
import seaborn as sb
from sklearn.datasets import load_iris 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris_ = pd.read_csv('iris.data')
columns = ['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm', 'class']
iris_.columns=columns
iris_.head()
```

```python
%matplotlib inline 
plt.figure(figsize=(20,10))
sb.pairplot(iris_.dropna(),hue='class')
```

```python
plt.figure(figsize=(20,10))
for index,column in enumerate(iris_.columns):
    if column == 'class':
        continue
    plt.subplot(2,2,index+1)
    sb.violinplot(x='class',y=column,data=iris_)
```
