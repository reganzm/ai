```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn import datasets
```

```python
#加载糖尿病数据
datas = datasets.load_diabetes()
x = datas.data
y = datas.target
```

```python
np.sum(datas.data[:,0]**2)
```

```python
datas.data
```

```python
#绘制10幅线性回归图表，分别标识一个生理因素跟糖尿病病情的相关性。
x_train = x[:-20]
y_train = y[:-20]
x_test = x[-20:]
y_test = y[-20:]

linreg = LinearRegression()
#
plt.figure(figsize=(20,20))
for i in range(0,10):
    xi_test = x_test[:,i]
    xi_train = x_train[:,i]
    #xi_test = x_test[:,np.newaxis]
    #xi_train = x_train[:,np.newaxis]
    model = linreg.fit(xi_train.reshape(-1, 1),y_train)
    y = model.predict(xi_test.reshape(-1, 1))
    plt.subplot(5,2,i+1)
    plt.scatter(xi_test,y_test)
    plt.plot(xi_test,y,color='b')
plt.show()
```
