```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.构建数据集x,y,满足y = 3.4x + 7.8 
```

```python
num = 500
datas = []
for i in range(num):
    x = np.random.normal(0,1)+1
    y = 0.2*x + 0.2 + np.random.normal(0,0.09)
    datas.append([x,y])
```

```python
#2.转换为numpy中的ndarray对象
x_data = [x[0] for x in datas]
y_data = [x[1] for x in datas]
```

```python
#3.datas数据可视化，绘制散点图
plt.figure(figsize=(20,10))
plt.scatter(x_data,y_data)
plt.show()
```

```python
#4.使用tensorflow来做线性回归
#迭代次数20
epoch = 10000
#随机初始化w,和b
w = tf.Variable(tf.random_normal([1],-0.5,0.5),name='W')
b = tf.Variable(tf.zeros([1]),name='b')
y = w * x_data + b

#使用梯度下降算法最优化loss值
loss = tf.reduce_mean(tf.square(y-y_data),name='loss')
optimizer = tf.train.GradientDescentOptimizer(0.001)
train = optimizer.minimize(loss,name='train')

init_op  = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init_op)
    print('initialize w:%s,b:%s,loss:%s'%(sess.run(w),sess.run(b),sess.run(loss)))
    for _ in range(epoch):
        sess.run(train)
        if(_ % 1000 ==0):
            print('initialize w:%s,b:%s,loss:%s'%(sess.run(w),sess.run(b),sess.run(loss)))

    #使用y值和x_data绘制拟合直线
    plt.figure(figsize=(20,10))
    plt.scatter(x_data,y_data)
    plt.plot(x_data,sess.run(w)*x_data+sess.run(b),'r-.')
    plt.show()
```
