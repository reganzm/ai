```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#L1正则化
def L1(w1,w2):
    return np.abs(w1)+np.abs(w2)
#L2正则化
def L2(w1,w2):
    return w1**2 + w2**2
```

```python
n = 256
x = np.linspace(-2,2,n)
y = np.linspace(-2,2,n)
```

## 生成网格数据

```python
X,Y = np.meshgrid(x,y)
```

```python
X
```

```python
Y
```

## 绘制L2正则化等高线图

```python
plt.contourf(X,Y,L2(X,Y),60,alpha=0.5,cmap=plt.cm.hot,)
plt.show()
```

```python
plt.contour(X, Y,L2(X,Y), 6, colors = 'black', linewidth = 0.2)
plt.show()
```

## 绘制L1正则化登高线图

```python
plt.contourf(X,Y,L1(X,Y),60,alpha=0.75,cmap=plt.cm.hot)
plt.show()
```
