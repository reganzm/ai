### æ–¹å·®è†¨èƒ€å› å­

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# æž„é€ ç©ºçš„æ•°æ®æ¡†
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```

```python
type(vif)
```

```python
filterd_vif = vif[vif['VIF Factor']<=10]
```

```python
filterd_vif
```

```python
feauture_list = filterd_vif['features'].values.tolist()
feauture_list.pop(0)
feauture_list
```

```python
len(feauture_list)
```

```python
train.columns
```

```python
feauture_list.append('prediction_pay_price')
```

```python
train_f1 = train[feauture_list]
```

### å¼‚å¸¸å€¼æ£€æŸ¥

```python
import statsmodels.api as sm
formula = 'prediction_pay_price'+"~"+"+".join(feauture_list)
fit = sm.formula.ols(formula,data=train_f1).fit()
fit.summary()
```

```python
#å‡æ–¹è¯¯å·®
from sklearn.metrics import mean_squared_error
import numpy as np
pred = fit.predict()
np.sqrt(mean_squared_error(train_f1.prediction_pay_price,pred))
```

### ç¦»ç¾¤ç‚¹æ£€æŸ¥

```python
#ç¦»ç¾¤ç‚¹æ£€æµ‹
outliers = fit.get_influence()
# é«˜æ æ†å€¼ç‚¹ï¼ˆå¸½å­çŸ©é˜µï¼‰
#leverage = outliers.hat_matrix_diag
# dffitså€¼
#dffits = outliers.dffits[0]
# å­¦ç”ŸåŒ–æ®‹å·®df
resid_stu = outliers.resid_studentized_external
# cookè·ç¦»
#cook = outliers.cooks_distance[0]
# covratioå€¼
#covratio = outliers.cov_ratio
# å°†ä¸Šé¢çš„å‡ ç§å¼‚å¸¸å€¼æ£€éªŒç»Ÿè®¡é‡ä¸ŽåŽŸå§‹æ•°æ®é›†åˆå¹¶
#contat1 = pd.concat([pd.Series(leverage, name = 'leverage'),pd.Series(dffits, name = 'dffits'),
#                     pd.Series(resid_stu,name = 'resid_stu'),pd.Series(cook, name = 'cook'),#
#                     pd.Series(covratio, name = 'covratio'),],axis = 1)
contat1 = pd.concat([pd.Series(resid_stu,name = 'resid_stu')],axis = 1)
df_outliers = pd.concat([train_f1,contat1], axis = 1)
df_outliers
```
